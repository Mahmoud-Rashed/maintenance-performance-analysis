```python
names_list = list(df['قام بالاصلاح 1'].unique()) +\
list(df['قام بالاصلاح 2'].unique()) +\
list(df['قام بالاصلاح 3'].unique()) +\
list(df['قام بالاصلاح 4'].unique())

names_set = set(names_list)
names_set.remove(np.nan)
names_set
names = list(names_set)
names

Recieved = []
Fixed = []
Fixed_first = []
Fixed_second = []
Fixed_third = []
Fixed_fourth = []
Team = []
Plant = []

print("\n*** عدد المشاركات فى الاصلاح ***\n")

for name in names:
  n1 = 0
  n2 = 0
  n3 = 0
  n4 = 0
  r1 = 0

  team = df_users.loc[df_users["Name"] == name, "Shift"]

  team = team.iloc[0] if not team.empty else "X"
  print(team)

  plant = df_users.loc[df_users["Name"] == name, "Plant"]

  plant = plant.iloc[0] if not plant.empty else "X"
  print(plant)




  if name in df["قام بالاصلاح 1"].value_counts():
    n1 = int(df["قام بالاصلاح 1"].value_counts()[name])
  if name in df["قام بالاصلاح 2"].value_counts():
    n2 = int(df["قام بالاصلاح 2"].value_counts()[name])
  if name in df["قام بالاصلاح 3"].value_counts():
    n3 = int(df["قام بالاصلاح 3"].value_counts()[name])
  if name in df["قام بالاصلاح 4"].value_counts():
    n4 = int(df["قام بالاصلاح 4"].value_counts()[name])
  if name in df["مستلم الاخطار"].value_counts():
    r1 = int(df["مستلم الاخطار"].value_counts()[name])
  Recieved.append(r1)
  Fixed.append(n1+n2+n3+n4)
  Fixed_first.append(n1)
  Fixed_second.append(n2)
  Fixed_third.append(n3)
  Fixed_fourth.append(n4)
  Team.append(team)
  Plant.append(plant)
  print( "(" + str(names.index(name)+1) + ")  " + name + "\n(" + str(r1) + ")  " +\
        str(n1) + "  " + str(n2) + "  "+ str(n3) + "  " + str(n4) + "  ("+ str(n1+n2+n3+n4) + ")\n")
maintenance_team = {"Name":names,"Recieved":Recieved,"Fixed":Fixed,"Fixed_first":Fixed_first,"Fixed_second":Fixed_second,
                    "Fixed_third":Fixed_third,"Fixed_fourth":Fixed_fourth,"Team":Team,"Plant":Plant}
```

```python
import arabic_reshaper
from bidi.algorithm import get_display
```

```python
def fix_arabic(text):
    reshaped = arabic_reshaper.reshape(text)
    return get_display(reshaped)
```

```python
plt.figure(figsize=(16, 5))

mdf["Name_fixed"] = mdf["Name"].apply(fix_arabic)
mdf.sort_values(by=['Fixed'], inplace=True, ascending=False)

df_plot = (
    mdf[mdf["Fixed"] > 10]
    .sort_values(by="Fixed", ascending=False)
)

ax = sns.barplot(
    data=df_plot,
    x="Name_fixed",
    y="Fixed",
    orient="x"
)

for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(
        f"{int(height)}",
        (bar.get_x() + bar.get_width() / 2, height),
        ha="center",
        va="bottom",
        xytext=(0, 1),
        textcoords="offset points",
        fontsize=9
    )

plt.xticks(rotation=90)
plt.show()
```
